Between-tissue Gene Content
---------------------------

Let's move on to a less naive assessment of gene content and expression. We want to get an idea of how all samples compare with each other.

For background, consider that each sample can be viewed as a vector of transcript abundances. We then need only choose a distance metric to compare two samples with each other, and group them by distance.

So, let's look at some different methods!

```python
from itertools import izip
from scipy.spatial.distance import pdist
from scipy.spatial.distance import squareform
import scipy.cluster.hierarchy as sch
```

```python
'''
Code for dendrogram plotting borrowed and modified from:
http://stackoverflow.com/questions/2982929/plotting-results-of-hierarchical-clustering-ontop-of-a-matrix-of-data-in-python/3011894#3011894
'''

def plot_dendro(df, distance='hamming', method='centroid', title_text='dendrogram', 
                cmap=plt.cm.Reds, fn='dendro_hamming.png', labels=sample_labels, figsize=(24,18)):
    
    dmat = squareform(pdist(df.transpose(), distance))
    
    sns.set(style='nogrid')
    fig = figure()
    fig.set_size_inches(*figsize);

    axdendro = fig.add_axes([0.07,0.1,0.14,0.8])
    Y = sch.linkage(dmat, method=method)
    Z = sch.dendrogram(Y, orientation='right')
    dendro_index = Z['leaves']
    
    ticklabels = [labels[col] for col in list(df.columns[dendro_index])]
    
    axdendro.set_frame_on(False);
    axdendro.set_xticks([]);
    axdendro.set_yticklabels(ticklabels, minor=False);

    axmatrix = fig.add_axes([0.3,0.1,0.6,0.8])
    dmat = dmat[dendro_index,:]
    dmat = dmat[:,dendro_index]
    im = axmatrix.matshow(dmat, aspect='auto', origin='lower', cmap=cmap)

    axmatrix.set_yticks([]);
    axmatrix.set_xticks(np.arange(len(df.columns)), minor=False);
    axmatrix.xaxis.tick_bottom()
    axmatrix.set_xticklabels(ticklabels, minor=False);
    axmatrix.set_title(title_text, size=24)
    plt.xticks(rotation=90);

    # Plot colorbar.
    axcolor = fig.add_axes([0.91,0.1,0.02,0.8])
    colorbar(im, cax=axcolor)
    savefig(fn)
    return fig
```

```python
def quantile_normalize(df):
    means = pd.DataFrame({col: df[col].order(ascending=False) for col in df}).mean(axis=1)
    return pd.DataFrame({col: means[np.argsort(np.array(-df[col]))] for col in df})
```

Hamming Distance
----------------

The first metric we'll look at is the Hamming distance. This does not take into account transcript abundance, but rather measures only transcript existence.

More formally, the Hamming distance here measures the number of disagreeing dimensions in the two vectors, normalized by the number of dimensions.

Mahalanobis Distance
--------------------

Mahalanobis distance normalizes by the covariance matrix; as such, it requires a PCA be done. Otherwise, we get a 700000 by 700000 element covariance matrix!

This is a work in progress...

```python
iso_PCA = PCA()
iso_PCA.fit(iso_rsem_df[blast_df.contaminants==False].transpose())
```
