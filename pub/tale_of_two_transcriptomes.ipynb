Intro
-----



A long time ago, in a lab really not all that far away, an unreasonable amount of RNA-seq was run on *Petromyzon marinus*, commonly called the sea lamprey.

The goals of this sequencing were broad, and the samples were taken in many different conditions, from many individuals.

By the time they were done, the number of reads had swelled into the billions. 

At the time, data of this size were impossible to assemble. Now, we have the techniques to do so.

Concerning Lampreys
-------------------

Lampreys are most commonly known for their role as fish parasites. 

```python
Image(url='http://upload.wikimedia.org/wikipedia/commons/2/28/Petromyzon_marinus.jpg')
# built in citation! IPython notebook is great!
```

They are an invasive species in the Great Lakes, and have had a devastating impact on the native fish population.

The impact of lampreys on the ecosystem and the economy of the mid-west has driven research.

Our collaborators in __Wieming Li's__ lab study the lamprey within the Fisheries and Wildlife Department.

```python
Image(url='http://msutoday.msu.edu/_/img/assets/2010/weiming-li-pnas.jpg', width=400)
```

There was also a concerted effort to sequence the lamprey genome, led by Jeremiah Smith of the University of Kentucky, the fruits of which were published in 2013.

```python
HTML('<iframe src=http://www.nature.com/ng/journal/v45/n4/full/ng.2568.html width=800 height=600></iframe>')
```

An important discovery was that lampreys undergo *somatic recombination*: somatic tissues drop large amounts of their DNA. This means that the genome, which was sequenced from liver DNA, is only about 70% complete. 

More Reasons to Care About Lampreys
-----------------------------------

* One of the oldest lineages of extant species, diverged from human ~500 mya
* Most recent common ancestor to the gnathostomes
* Whole genome duplications

Project Goals
-------------

* Show that assembling huge RNA-seq datasets de novo is tractable
* Demonstrate the scalability of our lab's k-mer based techniques to large transcriptomic data
* Show that co-assembly of many samples increases recall
* Generate a scientifically valuable set of lamprey transcripts

*My goals*: learn about RNA-seq, assembly algorithms and analysis, and some lamprey biology. And, do cool science!

The Pipeline
------------

The pineline is largely similar to that Eel Pond protocol -- the lamprey work helped give the go-ahead to develop the protocols.

```python
HTML('<iframe src=https://khmer-protocols.readthedocs.org/en/v0.8.4/mrnaseq/ width=800 height=4
     00></iframe>')
```

* First, we quality control with trimmomatic to remove adapters and low-quality sequence
* Next, we run digital normalization to a coverage of 50.
* Then, we run abundance trimming
* Run digital normalization again, to a coverage of 20
* Assemble with trinity

Homologies
----------

Homology with a well-studied species (or even better, the genome!) is a fair indicator of transcript integrity.

In our case, the incompleteness of the genome means it can't be used to validate *every* transcript, but a genome hit is still an excellent indicator for those transcripts the genome includes.

Let's step through some overall homology results:

```python
float(xgenome_ix.sum()) / len(iso_support_df) * 100.0
```

In total, about 62% of all transcripts have hits on the genome. Estimated genome completeness is ~70%.

```python
float((xgenome_ix | xmouse_ix).sum()) / len(iso_support_df) * 100.0
```

When we include mouse, that gets up to ~68.6%.

```python
float((xgenome_ix | xmouse_ix | xzebrafish_ix | xamph_ix).sum()) / len(iso_support_df) * 100.0
```

We only get a tiny jump, up to ~68.9%, when we add in zebrafish and amphioxus searches to mouse and genome.

Contamination?
--------------

A concern with the assembly has been the potential for contamination; there are lots of samples, introducing many opportunities for human error.

We have a *lot* of transcripts unique to the spinal cord with no support in existing resources. That raises some red flags.

So, let's look at homologies with human cdna.

```python
pd.DataFrame({'transcripts': iso_support_df.sum(),
         'hits': iso_support_df.loc[(xhuman_ix | xhuman_dna_ix),:].sum(),
         }).plot(kind='bar')
title('human homologies per sample')
savefig('human_hits_per_sample.png')
```

```python
Image('human_hits_per_sample.png')
```

The spinal cord sample looks suspect --

All other samples have a similar number of hits, likely core genes, but spinal cord has almost 50000!

Let's see how many of these transcipts with human homology are *not* in the genome.

```python
pd.DataFrame({'human hits': iso_support_df.loc[xhuman_ix | xhuman_dna_ix,:].sum(),
              'human unique dna': iso_support_df.loc[xhuman_dna_ix & ((xgenome_ix | xlamp0_ix | xEST_ix) == False),:].sum(),
         'human unique cdna': iso_support_df.loc[xhuman_ix & ((xgenome_ix | xlamp0_ix | xEST_ix) == False),:].sum(),
         }).plot(kind='bar');
title('human unique homologies per sample')
savefig('human_unique_hits_per_sample.png')
```

```python
Image('human_unique_hits_per_sample.png')
```

It seems quite clear that the spinal cord sample is contaminated with human cDNA and DNA. We can safely say that the spinal cord sample should not be trusted.

```python
((xhuman_ix | xhuman_dna_ix) & ((xgenome_ix | xlamp0_ix | xEST_ix) == False)).sum()
```

That's 41317 transcripts we can remove from the assembly.

```python
pd.DataFrame(index=((xhuman_ix | xhuman_dna_ix) & ((xgenome_ix | xlamp0_ix | xEST_ix) == False)).index).to_csv('human_contiminated.txt', delimiter=' ', header=False, index=True)
```
