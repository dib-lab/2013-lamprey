```python
def build_labels(labels):
    return ':'.join(['`{l}`'.format(l=l) for l in labels])
```

```python
def build_properties(props):
    return ', '.join(['{k}:{v:f}'.format(k=k, v=v) for k, v in props.iteritems()])
```

```python
def load_transcripts(df, labels=['lamp03', 'transcript', 'tax7757'], block_size=10000):
    for block_start in range(0, len(df), block_size):
        
        print 'processing block at', block_start
        
        tx = rdb.transaction(for_query=True)
    
        for n, (tr_name, row) in enumerate(df.ix[block_start:block_start+block_size].iterrows()):
            
            length = len(assem_db[tr_name].sequence)
            q = "MERGE (n:{labels} {{ name:'{name}', length={l}}})".format(
                                                                labels=build_labels(labels), 
                                                                name=tr_name, l=length)
            #tx.append(from_q)
            print q

            if n > 0:
                break
                
        print '...committing block at', block_start
        #tx.commit()
```

```python
def load_tissues(df, tissues):
    
    #tx = rdb.transaction(for_query=True)
    for sample, row in tissues.iterrows():
        q = "MERGE (n:{labels} {{ sample:'{s}', tissue:'{t}'}})".format(
                                                        labels=build_labels(['tissue', 'lamp03']), 
                                                        s=row.newname, t=row.tissue)
        print q
        #tx.append(q)
    #tx.commit()
    
    for block_start in range(0, len(df), block_size):
        
        print 'processing block at', block_start
        
        tx = rdb.transaction(for_query=True)
    
        for n, (tr_name, row) in enumerate(df.ix[block_start:block_start+block_size].iterrows()):
            
            for key, value in row.iteritems():
                if type(value) is not float:
                    q = textwrap.dedent('''\
                        MATCH (tr:{from_labels}), (tis:{to_labels})
                        WHERE tr.name = '{tr_name}' AND tis.sample = '{sample}'
                        CREATE UNIQUE (tr)-[r:`EXPRESSED_IN` {{ rsem:{rsem} }}]->(tis)
                '''.format(tr_name=tr_name, sample=key, rsem=value,
                           from_labels=build_labels(['lamp03', 'transcript']), 
                           to_labels=build_labels(['tissue', 'lamp03'])))
                
                    print q
                    
            #tx.append(from_q)
            print q

            if n > 0:
                break
                
        print '...committing block at', block_start
        #tx.commit()
```

```python
def load_blast(df, from_labels=['lamp03', 'transcript', 'tax7757'],
               to_labels=['protein'], block_size=10000):

    for block_start in range(0, len(df), block_size):

        print 'processing block at', block_start

        #tx = rdb.transaction(for_query=True)

        for n, (tr_name, row) in enumerate(df.ix[block_start:block_start+block_size].iterrows()):

            from_q = query_get_create_sequence(tr_name, labels=from_labels)
            #tx.append(from_q)

            print from_q
            
            if type(row.sseqid) is not float:
                pr_name = row.sseqid

                to_q = query_get_create_sequence(pr_name, labels=to_labels)
                #tx.append(to_q)
                print to_q
                about = row.to_dict()
                del about['sseqid']
                hom_info = build_properties(about)

                q = textwrap.dedent('''\
                MATCH (tr:{from_labels}), (pr:{to_labels})
                WHERE tr.name = '{tr_name}' AND pr.name = '{pr_name}'
                CREATE UNIQUE (tr)-[r:`HOMOLOGOUS` {{ {hom_info} }}]->(pr)
                '''.format(tr_name=tr_name, pr_name=pr_name, hom_info=hom_info,
                           from_labels=build_labels(from_labels),
                           to_labels=build_labels(to_labels)))
                print q
                #tx.append(q)
            if n > 0:
                break
        print '...committing block at', block_start
        #tx.commit()
```
