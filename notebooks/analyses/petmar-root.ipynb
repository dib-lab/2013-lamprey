### Transcript Support

````python
genes_rsem_df = pd.read_csv(wdir('all.lamp3.rsem.matrix'), delimiter='\t', index_col=0)
iso_rsem_df = pd.read_csv(wdir('all.lamp3.iso.rsem.matrix'), delimiter='\t', index_col=0)
````

````python
genes_support_df = pd.DataFrame(index=genes_rsem_df.index)
iso_support_df = pd.DataFrame(index=iso_rsem_df.index)
````

````python
for col in iso_rsem_df:
    iso_support_df[col] = 0
    iso_support_df.loc[iso_rsem_df[col] > 1.0, col] = 1
````

````python
for col in genes_rsem_df:
    genes_support_df[col] = 0
    genes_support_df.loc[genes_rsem_df[col] > 1.0, col] = 1
````

````python
labels = {k:v['label'] for k, v in metadata['samples'].items()}
````

````python
iso_support_df.rename(columns=labels, inplace=True)
iso_rsem_df.rename(columns=labels, inplace=True)
genes_support_df.rename(columns=labels, inplace=True)
genes_rsem_df.rename(columns=labels, inplace=True)

iso_support_df.sort_index(axis=1, inplace=True)
iso_rsem_df.sort_index(axis=1, inplace=True)
genes_support_df.sort_index(axis=1, inplace=True)
genes_rsem_df.sort_index(axis=1, inplace=True)
````

````python
labels_df = pd.DataFrame(labels, index=['newname']).transpose()
labels_df['tissue'] = labels_df.newname.apply(lambda x: x.partition('_')[0])
labels_df['tissue_c'] = Categorical(list(labels_df.tissue)).labels
labels_df.sort(columns='newname', inplace=True)
tissue_palette = sns.color_palette('Set3', n_colors=len(labels_df.groupby('tissue_c').count()))
labels_df['color'] = labels_df.tissue_c.apply(lambda x: tissue_palette[x])
````

We know this sample is contaminated; so, we'll mask any transcripts unique to that sample.

````python
contaminants = (iso_support_df.sum(axis=1) == 1) & iso_support_df.liver_3102H
````

````python
iso_support_df = iso_support_df[contaminants == False]
iso_rsem_df = iso_rsem_df[contaminants == False]
````

````python
store['iso_support_df'] = iso_support_df
store['iso_rsem_df'] = iso_rsem_df
store['labels_df'] = labels_df
````

### Blast Results

````python
blast_items = [(n, info) for n, info in metadata['urls'].items() + metadata['queries'].items() if n != 'assembly']
for i, (dbname, info) in enumerate(blast_items):
    target = wdir('{}.fasta.x.{}.db.tsv'.format(metadata['prefix'], info['dest'].rstrip('.gz')))
    print dbname, target
    tmp = blast.blast_to_df(target)
    blast.best_hits(tmp)
    if i == 0:
        blast_panel = pd.Panel({dbname: tmp})
    else:
        blast_panel[dbname] = tmp
````

````python
blast_items = [(n, info) for n, info in metadata['urls'].items() + metadata['queries'].items() if n != 'assembly']
for i, (dbname, info) in enumerate(blast_items):
    A_fn = '{}.fasta.x.{}.db.tsv'.format(metadata['prefix'], info['dest'].rstrip('.gz'))
    B_fn = '{}.db.x.{}.fasta.tsv'.format(info['dest'].rstrip('.gz'), metadata['prefix'])
    
    print '{} <=> {}'.format(A_fn, B_fn)
    
    A = blast.blast_to_df(wdir(A_fn))
    B = blast.blast_to_df(wdir(B_fn))

    X = blast.get_orthologies(A, B, iso_support_df.index)
    
    if i == 0:
        ortho_panel = pd.Panel({dbname: X})
    else:
        ortho_panel[dbname] = X
````

````python
blast_df = blast_panel.minor_xs('evalue') >= 0
````

````python
ortho_df = ortho_panel.minor_xs('evalue_x') >= 0
````

````python
putative_df = pd.DataFrame({'homology': (blast_df.sum(axis=1) > 0)})
putative_df['orthology'] = (ortho_df.sum(axis=1) > 0)
putative_df['not_lamp00'] = ((blast_df.petMar2_cdna | blast_df.petMar2_cds | \
                             blast_df.petMar2_pep | blast_df.petMar2_ncrna) == False)
putative_df['novel'] = ((blast_df.petMar2 == False) & putative_df.not_lamp00)
````

````python
tissue_tr_df = iso_support_df.groupby(by=labels_df.tissue.values, axis=1).sum()
tissue_support_df = iso_support_df.groupby(by=labels_df.tissue.values, axis=1).sum() > 0
````

````python
store['tissue_tr_df'] = tissue_tr_df
store['tissue_support_df'] = tissue_support_df
````

````python
store['blast_df'] = blast_df
store['ortho_df'] = ortho_df
````

````python
store['blast_panel'] = blast_panel
store['ortho_panel'] = ortho_panel
````

````python
store['putative_df'] = putative_df
````
