```python
import pandas as pd
import os
```

```python
import peasoup as ps
```

```python
def wdir(fn):
    return os.path.join('../../work/', fn)
```

```python
outfmt6 = ['sseqid', 'pident', 'length', 'mismatch', 'gapopen', 'qstart', \
           'qend', 'sstart', 'send', 'evalue', 'bitscore']
```

```python
A = [h for h in ps.blast.iter_distinct_csv('../../work/lamp03.fasta.x.musMus.pep.fa.db.tsv', header=outfmt6)]
```

```python
A = pd.read_table(wdir('lamp03.fasta.x.musMus.pep.fa.db.tsv'),
                    header=None, index_col=0, names=outfmt6)
A['ind'] = A.index
A.sort(columns=['ind', 'evalue'])
A.drop_duplicates(cols='ind', inplace=True)
del A['ind']
```

```python
B = pd.read_table(wdir('musMus.pep.fa.db.x.lamp03.fasta.tsv'),
                    header=None, index_col=0, names=outfmt6)
B['ind'] = B.index
B.sort(columns=['ind', 'evalue'])
B.drop_duplicates(cols='ind', inplace=True)
del B['ind']
```

```python
X = pd.merge(A, B, how='inner', left_on='sseqid', right_index=True)
```
