````python
def run_tasks(tasks, n_cpus=1):
    
    if type(tasks) is not list:
        raise TypeError('tasks must be a list')
    if n_cpus < 1:
        raise ValueError('n_cpus must be non-negative')
    if type(n_cpus) is not int:
        raise TypeError('n_cpus must be int')
    
    class Loader(TaskLoader):
        @staticmethod
        def load_tasks(cmd, opt_values, pos_args):
            config = {'verbosity': 2}
            return tasks, config
    cmd = []
    cmd.append('-n {}'.format(n_cpus))
        
    DoitMain(Loader()).run(cmd)
    
def apply_run_task(df, task_func):
    
    task_list = list(df.apply(task_func, axis=1, reduce=False))
    run_tasks(task_list)
````

````python
def get_task_files_df(tasks):
    targets = []
    for task in tasks:
        for target in task.targets:
            for fd in task.file_dep:
                targets.append({'dep': fd, 'target': target})
    return pd.DataFrame(targets)
````
