```python
def testScaffolder(blast_file = 'x.genome.csv'):
    '''
    parser = argparse.ArgumentParser()
    parser.add_argument('-i', '--blast_file', dest='blast_file')
    parser.add_argument('-m', '--min_length', dest='min_length', type=int)
    parser.add_argument('-n', '--max_overlap', dest='max_overlap', type=int)
    parser.add_argument('-o', '--output', dest='output')
    parser.add_argument('-v', '--verbose', action='store_true', default=False)
    args = parser.parse_args()
    '''
    min_length = 50
    max_overlap = 10
    output = 'scaffold/lamp3.scaffold.test'
    verbose = False

    hits = hb.BlastDB(blast_file)
    hits.filterByLength(min_length)
    byQuery = hits.partitionByQuery(min_size=2)
    
    clusters = []
    dedupestats = []
    print "Running through aggregated queries..."
    for n, key in enumerate(byQuery.keys()):
        if n % 10000 == 0:
            print "Processed {} query groups...".format(n)
        # removes duplicate subjects by choosing the highest scoring hit
        qset = byQuery[key]
        rawSize = len(qset)
        #print qset
        nr = removeSubjectDupes(qset)
        del qset
        subdedupeSize = len(nr)
        #print nr
        # checks query ranges and finds the set allowing
        # the highest number of hits
        maxNr = simpleOverlap(nr, max_overlap)
        del nr
        maxnrSize = len(maxNr)
        # we only care if the query has > 1 hit
        if maxNr and len(maxNr) > 1:
            maxNr.sort(key=lambda hit: hit.qstart)
            clusters.append(maxNr)
            stats = [rawSize, subdedupeSize, maxnrSize]
            dedupestats.append(stats)
            del maxNr
    
    endDedupe = map(statsFunc, dedupestats)
    avgDedupe = math.fsum(endDedupe) / float(len(endDedupe))
    print "Cluster size reduced to {} on average\n".format(avgDedupe)
    
    # remove duplicate clusters
    # for now just ignore duplicated contigs; in the future,
    # use information to reconcile and form larger scaffolds
    n = 0
    for c in clusters:
        n += len(c)
    
    print "Number of clusters: ", len(clusters)
    print "...comprising {} contigs".format(n)
    print "### Writing output to {} ###".format(output)
    with open(output, 'wb') as ofp:
        for cluster in clusters:
            if verbose:
                print '*' * 40
                for h in cluster:
                    print h.query, h.subject, h.bitscore, h.evalue
                print '*' * 40
            query = cluster[0].query
            subs = []
            for h in cluster:
                subs.append(h.subject)
                subs.extend([str(h.qstart), str(h.qend), str(h.sstart), str(h.send)])
            row = query + ':' + ','.join(subs)
            ofp.write(row + '\n')
    print "Done!"

    return clusters
```

```python
def overlap(hits, max_overlap):
    if hits and len(hits) > 1:
        H = sorted(hits, key=lambda hit:hit.qend)
        H_new = [H[0]]
        for i in range(1, len(H)):
            if not overLap(H[i], H_new[len(H_new)-1], max_overlap):
                H_new.append(H[i])
        return H_new
    return []

class ScaffoldEdge:
    
    def __init__(self, src, dst):
        self.src_hit = src
        self.dst_hit = dst
        
        self.src_name = src.subject
        self.dst_name = dst.subject
        
        self.transcript = src.query
    
    def __repr__(self):
        return "[{} {}:{} {}:{}]".format(self.transcript, self.src_hit.qstart, self.src_hit.qend, self.dst_hit.qstart, self.dst_hit.qend)
    
    def __str__(self):
        return "[{} {}:{} {}:{}]".format(self.transcript, self.src_hit.qstart, self.src_hit.qend, self.dst_hit.qstart, self.dst_hit.qend)

def build_mrna_graph(blast_file, min_overlap=10, min_length=50, max_evalue=1e-20, min_bitscore=50, verbose=False, limit=100):
    
    too_small = 0
    overlaps = 0
    edges = 0
    
    G = nx.DiGraph()
    with open(blast_file, 'rb') as bfp:
        for n, cluster in enumerate(hb.BlastParse.iter_csv_byquery(bfp, min_length=min_length, max_evalue=max_evalue, min_bitscore=min_bitscore)):
            if n % 10000 == 0:
                print "{} transcript clusters processed...".format(n)
                print cluster
            cluster = hb.BlastParse.dedupe(cluster, by='subject')
            if limit and n > limit:
                break
            elif len(cluster) < 2:
                too_small += 1
                if verbose:
                    print "Cluster < 2 subjects:", cluster
            elif len(cluster) > len(overlap(cluster, min_overlap)):
                overlaps += 1
                if verbose:
                    print "Cluster has overlapping conflicting hits:"
                    print cluster
                    print overlap(cluster, min_overlap)
            else:
                
                for i in range(0, len(cluster)-1):
                    for j in range(i+1, len(cluster)):
                        edges += 1
                        src = cluster[i]
                        dst = cluster[j]
                        edge = ScaffoldEdge(src,dst)
                        G.add_edge(src.subject, dst.subject, object=edge)
                        if verbose:
                            print "[{} {}:{}] --> [{} {}:{} {}:{}] --> [{} {}:{}]".format(src.subject, src.sstart, src.send, \
                                                                    edge.transcript, src.qstart, src.qend, \
                                                                    dst.qstart, dst.qend, \
                                                                    dst.subject, dst.sstart, dst.send) 
    print n
    return G, too_small, overlaps, edges

def clip_graph(G):
    
    print "Clipping "
    num_pruned=0
    for n in G.nodes():
        if G.out_degree(n) == 0 and G.in_degree(n) == 1:
            G.remove_node(n)
            #print "removed node", n
            num_pruned += 1
    for n in G.nodes():
        if G.out_degree(n) == 0 and G.in_degree(n) == 0:
            G.remove_node(n)
            num_pruned += 1
    print "...pruned", num_pruned, "nodes"


def remove_acyclic_components(G):
    print "iterating through components..."
    num_acyclic = 0
    num_cyclic = 0
    for comp in nx.weakly_connected_component_subgraphs(G):
        if nx.is_directed_acyclic_graph(comp) == False:
            print "found cyclic component! removing nodes..."
            num_cyclic += 1
            print comp.edges()
            print nx.simple_cycles(comp)
            G.remove_nodes_from(comp.nodes())
            G.remove_edges_from(comp.edges())
            print "removed {} nodes and {} edges".format(len(comp.nodes()), len(comp.edges()))
        else:
            num_acyclic += 1



```

```python
from itertools import groupby

def getDir(hit):
    qdir = 0
    sdir = 0
    if hit.qstart > hit.qend:
        qdir = 1
    if hit.sstart > hit.send:
        sdir =1
    return (qdir, sdir)

def dirVal(aSeq):
    skey = lambda hit: hit.subject
    hits = sorted(aSeq.hits, key=skey)
    if len(aSeq.hits) > 1:
        for k, g in groupby(hits, skey):
            subHits = [s for s in g]
            D = getDir(subHits[0])
            for hit in subHits[1:]:
               if getDir(hit) != D:
                   return False
    return True


def validateTranscripts(blast_pSet):
    valid = 0
    invalid = 0
    seqCounter = 0
    
    new_pSet = hb.PartitionSet()
    
    print "Starting validation process: checking {} transcripts".format(blast_pSet.numSeqs())
    for n, partition in enumerate(blast_pSet):
        for aSeq in partition:
            if seqCounter % 50000 == 0:
                print "{} partitions, {} sequences checked...".format(n, seqCounter)
                print "{} valid, {} invalid so far...".format(str(valid), str(invalid))
            seqCounter += 1
            if dirVal(aSeq):
                new_pSet.addSeq(aSeq, partition.pid)
                valid += 1
            else:
                invalid += 1
    return valid, invalid, new_pSet
    
```
