**k-mer abundance**

K-mer abundances for the lump and non-lump regions. Distributions are similar, though we see that in the lump, the peak is *slightly* shifted to the right, and more importantly, there is a higher density of higher-abundance k-mers, especially beyond 3-4.

```python
plot(lump_dist['abund'], lump_dist['count'], label='lump', lw=3.0)
plot(nlump_dist['abund'], nlump_dist['count'], label='non-lump', lw=3.0)
title('k-mer abundance dists')
xlabel('abundance')
ylabel('abundance')
legend()
axis(xmax=20)
```

**Abundance by position percentile**

k-mer abundance by percentile position in transcript. Nothing terribly exciting here; the expected lower k-mer abundance at the ends of transcripts, increasing in the center. We see that the lump region has a higher abundance across the transcript, and as expected, a more noticable $$\mathrm{d}abund$$
