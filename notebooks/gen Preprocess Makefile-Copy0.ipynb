```python
makefile = '''
KHMER=~/bin/khmer
SEQCLEAN=seqclean
TRIM=~/bin/Trimmomatic-0.25/trimmomatic-0.25.jar
TIME=/usr/bin/time -f %E%U%K%W%w%I%O
KMERGENIE=~/bin/kmergenie-1.5397/kmergenie

KMERRES=nodes=1:ppn=4,walltime=96:00:00,mem=32gb
TRIMRES=nodes=1:ppn=4,walltime=12:00:00,mem=4gb
QCRES=nodes=1:ppn=1,walltime=2:00:00,mem=4gb
DGRES=nodes=1:ppn=1,walltime=72:00:00,mem=42gb

ABYSSRES=nodes=1:ppn=24,walltime=3:00:00:00,mem=384gb
MIRARES=nodes=1:ppn=16,walltime=7:00:00:00,mem744gb
'''
```

```python
fqc1 = []
for f in reads:
    name = os.path.basename(f)
    bname, _, _ = name.partition('.')  
    with open('{}.fastqc1.sub'.format(bname), 'wb') as pbsfp:
        sub = '''
#!/bin/bash -login
cd $PBS_O_WORKDIR
module load FastQC
make {}.fastqc1'''.format(bname)
        
        pbsfp.write(sub)
        
        rule = '''
{bname}.fastqc1:
\tfastqc --noextract {name}
'''.format(bname=bname, name=name)
        
        pbsrule = '''
{bname}.fastqc1-pbs:
\tqsub -l $(QCRES) {bname}.fastqc1.sub
'''.format(bname=bname)
        
        makefile += rule
        makefile += pbsrule
        
        fqc1.append('{bname}.fastqc1-pbs'.format(bname=bname))

makefile += '\nfastqc1-pbs: {}'.format(' '.join(fqc1))
```

```python
# \tln -s {name} {src_dir}/{name}
   
trimmed = []
for f in reads:
    name = os.path.basename(f)
    bname, _, _ = name.partition('.')  
    with open('{}.trim.sub'.format(bname), 'wb') as pbsfp:
        sub = '''
#!/bin/bash -login
cd $PBS_O_WORKDIR
make {}.trim.fq'''.format(bname)
        
        pbsfp.write(sub)
        
        rule = '''
{bname}.trim.fq: {name}
\t\t$(TIME) -o {bname}.trim.time java -classpath $(TRIM) org.usadellab.trimmomatic.TrimmomaticSE -trimlog {bname}.trimlog -threads 4 {name} {bname}.trim.fq ILLUMINACLIP:adapters.fa:2:40:15 SLIDINGWINDOW:4:15 MINLEN:31
'''.format(bname=bname, name=name)
        
        pbsrule = '''
{bname}.trim.fq-pbs:
\tqsub -l $(TRIMRES) {bname}.trim.sub
'''.format(bname=bname)
        
        makefile += rule
        makefile += pbsrule
        
        trimmed.append('{bname}.trim.fq-pbs'.format(bname=bname))

makefile += '\ntrim-pbs: {}'.format(' '.join(trimmed))
```
